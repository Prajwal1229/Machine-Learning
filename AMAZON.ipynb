```python
# Lets  Remove Unnecessary Characters & Lowercase Text
# Lets import Regular expression
```

```python
import re
```

```python
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text
```

```python
df['Cleaned_Text'] = df['Text_Review'].apply(clean_text)
```

```python
# Convert Text into Numerical Data (Feature Extraction)Since models cant understand raw text, we need to convert it into vectors.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=5000)  # Limit to 5000 most important words
X = vectorizer.fit_transform(df['Cleaned_Text'])  # Transform text into numerical vectors
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Since we are starting with Logistic Regression lets build a model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Model=LogisticRegression(class_weight={0:2, 1:4, 2:1}, random_state=42)
```

```python
Model.fit(x_train,y_train)
```

```python
Model.score(x_train,y_train)
```

```python
Model.score(x_test,y_test)
```

### Since our model is low bias and low Varience its Genralized Fitting

```python
# Hear we are importing the classification report and confusion matrix 
# classification report--The classification report provides detailed metrics for each class in the dataset
# confusion matrix--it is a table that summarizes the performance of a classification model by comparing predicted labels with true labels.
from sklearn.metrics import confusion_matrix,classification_report
```

```python
y_train_predict=Model.predict(x_train)
y_test_predict=Model.predict(x_test)
```

```python
# Now lets create a Heat map to check the performance of a classification model by comparing predicted labels with true labels 
sns.heatmap(confusion_matrix(y_train,y_train_predict),annot=True,fmt='g',cmap='Blues');
```

```python
# Now lets print classification report so it provides detailed metrics for each class in the dataset
print(classification_report(y_train,y_train_predict))
```

```python
print(classification_report (y_test,y_test_predict))
```
